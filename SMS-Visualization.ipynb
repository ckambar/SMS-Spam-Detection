```python
#import the required modules
import numpy as np
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time

class Visualization:
    def __init__(self):
        self.SpamFilter, self.HamFilter = self.Data_Cleanup()
        
    def Data_Cleanup(self):
        #Read the dataset as DataFrame
        dataset=pd.read_csv("spam.csv", encoding="latin-1")
        dataset=dataset.drop(["Unnamed: 0"], 1)

        #Rename the column names
        dataset = dataset.rename(columns={"v1":"label", "v2":"message"})

        df = pd.DataFrame(dataset)

        #Seperation of spam messages from entire dataset
        SpamFilter = df[df['label']=='spam']

        #Seperation of ham messages from entire dataset
        HamFilter = df[df['label']=='ham']

        return SpamFilter, HamFilter

    def Visualize_Spam(self):
        #Visualize what words appear the most in spam messages

        wc=WordCloud(background_color='White', width=1000, height=1000,max_words=200).generate(str(self.SpamFilter['message']))
        plt.rcParams['figure.figsize']=(10,10)
        plt.title("Most common words in spam messages")
        plt.axis('off')
        plt.imshow(wc)
    
    def Visualize_Ham(self):
        #Visualize what words appear the most in ham messages

        wc=WordCloud(background_color='White', width=1000, height=1000,max_words=100).generate(str(self.HamFilter['message']))
        plt.rcParams['figure.figsize']=(10,10)
        plt.title("Most common words in ham messages")
        plt.axis('off')
        plt.imshow(wc)
```

```python
data_Visualization=Visualization()
```

```python
data_Visualization.Visualize_Spam()
```

```python
data_Visualization.Visualize_Ham()
```
